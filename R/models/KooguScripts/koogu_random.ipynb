```python
import os
from koogu.data import preprocess, feeder
from koogu.model import architectures
from koogu import train, recognize
from matplotlib import pyplot as plt

# Path to the parent directory containing multiple directories of training files
parent_training_dir = '/Volumes/DJC Files/JahooGibbonClipsRandomSmall/'

# List directories to process
training_dirs = [os.path.join(parent_training_dir, d) for d in os.listdir(parent_training_dir) if os.path.isdir(os.path.join(parent_training_dir, d))]

# Function to extract sample count from directory name
def extract_sample_count(dir_name):
    parts = dir_name.split('_')
    for part in parts:
        if part.endswith('samples'):
            return int(part.replace('samples', ''))
    return float('inf')  # Default to a very high number if the pattern is not found

# Sort directories based on sample count
training_dirs.sort(key=lambda d: extract_sample_count(os.path.basename(d.rstrip('/'))))

data_settings = {
    'audio_settings': {
        'clip_length': 12.0,
        'clip_advance': 0.4,
        'desired_fs': 32000
    },
    'spec_settings': {
        'win_len': 0.128,
        'win_overlap_prc': 0.75,
        'bandwidth_clip': [500,3000]
    }
}

output_root_base = '/Volumes/DJC Files/koogu_benchmark/data/'

# Process each training directory
for training_dir in training_dirs:
    # Extract the directory name to use as part of the output path
    dir_name = os.path.basename(training_dir.rstrip('/'))
    output_root = os.path.join(output_root_base, dir_name)

    # Convert audio files into prepared data
    clip_counts = preprocess.from_top_level_dirs(
        data_settings['audio_settings'],
        class_dirs=['gibbon', 'noise'],
        audio_root=training_dir,
        output_root=output_root,
        negative_class_label='noise'
    )
    print(f"Processed {training_dir}: {clip_counts}")

    # Create data feeder for the current directory
    data_feeder = feeder.SpectralDataFeeder(
        output_root,
        data_settings['audio_settings']['desired_fs'],
        data_settings['spec_settings'],
        validation_split=0.15,
        max_clips_per_class=20000
    )

    print(data_feeder)

    model = architectures.DenseNet(
        [4, 4, 4],
        preproc=[('Conv2D', {'filters': 16})],
        dense_layers=[32]
    )

    training_settings = {
        'batch_size': 64,
        'epochs': 100,
        'learning_rate': 0.01,
        'lr_change_at_epochs': [20, 40],
        'lr_update_factors': [1.0, 1e-1, 1e-2],
        'dropout_rate': 0.05
    }

    output_root_model = '/Volumes/DJC Files/koogu_benchmark/model/'
    model_dir = os.path.join(output_root_model, dir_name)

    # Perform training
    history = train(
        data_feeder,
        model_dir,
        data_settings,
        model,
        training_settings
    )

    fig, ax = plt.subplots(2, sharex=True, figsize=(12, 9))
    ax[0].plot(
        history['train_epochs'], history['binary_accuracy'], 'r',
        history['eval_epochs'], history['val_binary_accuracy'], 'g')
    ax[0].set_ylabel('Accuracy')
    ax[1].plot(
        history['train_epochs'], history['loss'], 'r',
        history['eval_epochs'], history['val_loss'], 'g')
    ax[1].set_yscale('log')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Loss')
    plt.show()

    test_audio_root = '/Volumes/DJC Files/MultiSpeciesTransferLearning/TestData/Vietnam/ValidationClipsVietnam/'
    raw_detections_root = '/Volumes/DJC Files/koogu_benchmark/detections/'
    output_path = os.path.join(raw_detections_root, dir_name)

    recognize(
      model_dir=model_dir,
      audio_root=test_audio_root,
      output_dir=output_path,
      batch_size=64,
      recursive=True,
      show_progress=True,
      #reject_class='noise',
      threshold=0.1,
    )
```
